# 1.Conceptual Understanding

a.In Natural Language Processing, the difference between “Love” and “love” comes down to case sensitivity. Unless the text is normalized, the system treats them as separate tokens. This means that “Love” and “love” would be counted differently, which can affect frequency analysis and model training. Converting all text to lowercase is a common step to ensure consistency.

b.If stopwords are not removed, they can introduce unnecessary noise into the dataset. Stopwords such as “the,” “is,” and “at” occur very frequently but usually do not add meaningful information. Their presence can increase the size of the vocabulary and distort frequency-based models, making them less efficient.

c.There are situations where removing stopwords can be harmful. In sentiment analysis, words like “not” or “no” are essential for understanding meaning. For example, the phrase “not happy” conveys a negative sentiment, but if “not” is removed, the phrase becomes “happy,” which completely changes the interpretation. In legal or medical text analysis, stopwords may carry structural meaning. For instance, “without liability” or “free of charge” would lose their intended meaning if words like “without” or “of” were removed.

d.The difference between stemming and lemmatization lies in how they reduce words to their base form. Stemming is a rule-based process that crudely chops off suffixes, often producing non-standard roots such as “studies” becoming “studi.” Lemmatization, on the other hand, uses linguistic rules and dictionaries to return the proper base form of a word, such as “studies” becoming “study.” Lemmatization is more accurate but computationally heavier, while stemming is faster but less precise.


# 2.Build Advanced Preprocessing Function

```python
import re

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|[\w\.-]+@[\w\.-]+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    tokens = text.split()
    tokens = [t for t in tokens if len(t) > 2 or t in ["no", "not"]]
    return tokens, " ".join(tokens)
```

# 3.Stress Testing

```python
sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for s in sample_inputs:
    tokens, cleaned = preprocess_text(s)
    print(f"Original Text: {s}")
    print(f"Cleaned Tokens: {tokens}")
    print(f"Cleaned Sentence: {cleaned}\n")
```

# 4.Token Analytics

```python
def token_analytics(text_list):
    results = []
    for s in text_list:
        tokens, cleaned = preprocess_text(s)
        total_tokens = len(tokens)
        unique_tokens = len(set(tokens))
        avg_length = sum(len(t) for t in tokens)/total_tokens if total_tokens else 0
        results.append({
            "original": s,
            "cleaned": cleaned,
            "total_tokens": total_tokens,
            "unique_tokens": unique_tokens,
            "avg_token_length": avg_length
        })
    return results

analytics = token_analytics(sample_inputs)
for r in analytics:
    print(r)
```

# Analysis Questions(task 4)

a.The sentence that had the most noise was “I absolutely looooved this product 😍😍.” This is because it contained repeated characters and emojis, which had to be cleaned during preprocessing. Compared to the other inputs, this sentence required the most normalization to become usable.

b.The sentence that retained the most meaningful tokens after cleaning was “I am not happy with this.” After preprocessing, the important words “not,” “happy,” “with,” and “this” were preserved. These tokens capture the sentiment of the sentence accurately, showing that the preprocessing function successfully removed noise while keeping the essential meaning intact.

# 5.Frequency Analysis

```python
from collections import Counter

all_tokens = []
for s in sample_inputs:
    tokens, _ = preprocess_text(s)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)
print("Top 10 most frequent words:", counter.most_common(10))
print("Top 5 least frequent words:", counter.most_common()[-5:])
```

# 6.Build Full Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []
    for s in text_list:
        tokens, cleaned = preprocess_text(s)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

pipeline_output = full_pipeline(sample_inputs)
print(pipeline_output)
```
